````python
# bool()
print(bool.__doc__)
items = [1, 0, '', 'six-seven!', None, [], {}, (), True, int(), 3.14]

print('-' * 50)
for item in items:
    print(f"Item {item} of type {type(item)} is evaluated (bool(item)) as {bool(item)}")
````

### 2.3 Letter E

````python
# enumerate()
print(enumerate.__doc__)

items = 'We will split this sentence into words'.split()

for index, item in enumerate(items, start=1):
    print(f"Item {index} -> {item}")
````

````python
# int()
print(int.__doc__)

print('-' * 50)
print(int(10))
# binary to decimal integer (from right to left: 0 * 2^0 + 1 * 2^1 = 0 + 2 = 2)
# binary combinations on each position: 0, 1 (2 possible values)
print(int('10', base=2))
print(bin(2))
# octal to decimal integer (0 * 8^0 + 1 * 8^1 = 0 + 8 = 8)
# octal combinations on each position: 0 - 7 (8 possible values)
print(int('10', base=8))
print(oct(8))
# hexadecimal to decimal integer (0 * 16^0 + 1 * 16^1 = 0 + 16 = 16)
# hexadecimal combinations on each position: 0 - 9 and a - f (16 possible values)
print(int('10', base=16))
print(hex(16))
````

### 2.7 Letter R

````python
# range()
print(range.__doc__)
print('-' * 50)
print(range(5))

# to 'materialize' those numbers we can cast (convert) range to a list
print(list(range(5)))
print(list(range(10, 101, 10)))
print(list(range(0, -21, -2)))
````

````python
# reversed()
print(reversed.__doc__)
# complicated!!!
print(''.join(list(reversed("Abracadabra!"))))
# it is easier to use slices
print("Abracadabra!"[::-1])
print(list(reversed([10, 5, 1])))
````

````python
# round()
print(round.__doc__)
print(1 / 3)
print(round(1 / 3))
print(round(1 / 3, 2))
print(round(1 / 3, 1)) 
````

### 2.8 Letter S

````python
# set()
print(set.__doc__)

print(set("Abracadabra"))
print(set(['Bob', 'Alice', 'Bob', 'Brian', 'John', 'Alice']))

usr_id = set(range(1, 15))
other_ids = set(range(10, 25))

usr_id, other_ids

# finding common items
print(usr_id.intersection(other_ids))
# or
print(usr_id & other_ids)
````

````python
# sorted()
print(sorted.__doc__)

print(sorted([3, 4, 7, 1, 3, 8, 9]))
print(sorted([3, 4, 7, 1, 3, 8, 9], reverse=True))
print(sorted(['Bob', 'Alice', 'Bob', 'Brian', 'John', 'Alice']))
# sorting by custom condition - here by the length of each string
print(sorted(['Bob', 'Alice', 'Bob', 'Brian', 'John', 'Alice'], key=len))

# how does it work?
# the key parameter takes a collable (let's say a function) and runs this function on each element of
# the sequence that is being sorted remembering the result of this function for each item
# then it sorts the items in the same order as the results (here of running len() function)
# this is very useful and does not require writing a lot of code!
````

````python
# sum()
print(sum.__doc__)
print('-' * 50)
print(sum(range(1, 101)))

print(sum(range(1, 11), start=100)) # start + sum of numbers
````

### 2.9 Letter Z

````python
# zip()
print(zip.__doc__)

print('-' * 50)
# works as a zipper imagine all the interables as columns of values and zip takes one element from
# each iterable and returns a tuple with those values
names = ['Alice', 'Alice', 'Bob', 'Bob', 'Brian', 'John']

print(list(zip(names, range(len(names)))))
print(list(zip(names, range(1, len(names) + 1))))
````
