## 1. Loading the data

Biblioteka pandas obsługuje wiele formatów danych, ale i sposobów ich wczytywania (np. bezpośrednio z Internetu poprzez protokół HTTP). 

Let's create a simple pandas DataFrame from a dictionary.

```python
my_data = {
    'name': ['Bob','Julia','Greg','John','Wolfgang','Maciek','Kate'],
    'score': [12, 10, 13, 9, 12, 14, 11]
}

# we are creating pandas dataframe from dictionary data, where each key becomes a column in the frame
df_1 = pd.DataFrame(my_data)
```

```python
df_1
```

As you see there are three columns where the first one was created automatically and is called an `index`.


Let's write the data to a file, and then load them once again.

```python
# pandas allows us to save it's data to a file with different formats
# let's choose something simple and very popular - CSV (comma separated values)
df_1.to_csv('my_data.csv')
```

Now you can see a file `my_data.csv` within the same folder as the notebook. You can also view it directly within web browser. You probably noticed that the column `index` is also present, which is not necessary somethimes. We can avoid this behaviour by providing a flag `index` set to `False` to do not save the index into a file. Let's try again overwriting the file.

```python
df_1.to_csv('my_data.csv', index=False)
```

Now the `index` column is gone. We can also do the same with the column headers, but this feature is rather a good default to store then within a file. We can set the `header=False` flag though to disable column header from being written into a file.

**Now lets read the data from a file.**

Let's load the data back again to practice.

```python
df_2 = pd.read_csv('my_data.csv')
df_2
```
