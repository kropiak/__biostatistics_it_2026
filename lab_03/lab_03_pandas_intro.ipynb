## 1. Loading the data

The pandas library supports many data formats, as well as various methods for loading them (e.g., directly from the Internet via the HTTP protocol).
 

Let's create a simple pandas DataFrame from a dictionary.

```python
my_data = {
    'name': ['Bob','Julia','Greg','John','Wolfgang','Maciek','Kate'],
    'score': [12, 10, 13, 9, 12, 14, 11]
}

# we are creating pandas dataframe from dictionary data, where each key becomes a column in the frame
df_1 = pd.DataFrame(my_data)
```

```python
df_1
```

As you see there are three columns where the first one was created automatically and is called an `index`.


Let's write the data to a file, and then load them once again.

```python
# pandas allows us to save it's data to a file with different formats
# let's choose something simple and very popular - CSV (comma separated values)
df_1.to_csv('my_data.csv')
```

Now you can see a file `my_data.csv` within the same folder as the notebook. You can also view it directly within web browser. You probably noticed that the column `index` is also present, which is not necessary somethimes. We can avoid this behaviour by providing a flag `index` set to `False` to do not save the index into a file. Let's try again overwriting the file.

```python
df_1.to_csv('my_data.csv', index=False)
```

Now the `index` column is gone. We can also do the same with the column headers, but this feature is rather a good default to store then within a file. We can set the `header=False` flag though to disable column header from being written into a file.

**Now lets read the data from a file.**

Let's load the data back again to practice.

## 3. Selecting data from pandas DataFrame

We can select any subset of data from the original DataFrame and there are many different ways to achieve the goal. 

### 3.1 Selecting by column/index oraz indices

```python
# single column - returns pandas Series
df_3['Country']
```

```python
# many columns - returns pandas DataFrame
df_3[['Country','Employee']]
```

```python
# selecting may be done by providing names of columns or values in index column (for row selection)
#  or by providing indices of those columns or rows

# loc method (index name, [columns])
# one row
df_3.loc[4, ['Employee']]
```

```python
# loc method (index name, [columns])
# rows with values from 0 to 5 and all columns (:) - slicing
df_3.loc[:5, :]
```

```python
# iloc uses indices instead of names

# slice - indexes 0,1,2,3,4 (stop = 5 is excluded)
df_3.iloc[:5, :]
```

### 3.2 Filtering data by condition

Filtering data is about creating a logical condition which is then applied on each given item in the data set oraz a slice from it.

This is actually a two step process. Creating a condition produces a `boolean map` (vector/matrix/tensor) with `True` and `False` values. Then this map in applied to real data and those which are at the positions where map contains `True` value are being selected, and those on positions with `False` value are being filtered out.

Let's try it out on an example.

```python
order_value_sample = df_3['Order_value'][:5]
order_value_sample
```

```python
# examplary map, created from a condition, separated from the data
order_value_sample > 1000
```

```python
# now, we apply that map (filter) on the data
order_value_sample[order_value_sample > 1000]
```

```python
# let's try it on sample with 2 columns
data_sample = df_3.loc[:4, ['Employee', 'Order_value']]
data_sample
```

```python
# the same filter, but on DataFrame, this will show whole rows, where in the given column the condition is met
# map is the same
print(data_sample['Order_value'] > 1000)

# but applied to rows
data_sample[data_sample['Order_value'] > 1000]
```

```python
# we can apply multiple conditions like filtering any table (in excel for example)
# [(condition) & (condition)] - sum of conditions (condition AND condition)
# [(condition) | (condition)] - alternative conditions (condition OR condition)
df_3[(df_3['Order_value'] > 5000) & (df_3['Country'] == 'Polska')]
```

One can also use functions to work with strings like in the examles below.

```python
# str.startswith - looks for a string at the beginning of a compared string
df_3[df_3['Order_date'].str.startswith('2004')]
```

```python
# in - it is a contains relation - answers a question if an object is in collection/squence
# and as we know Python str type is a sequence - we ca check if a string is within a strind
'Alice' in 'Alice in wonderland'
```

```python
# but in pandas array (numpy is being used as a data library for it) we need to do it in another way
# we will use a str.contains() function
```

```python
df_3[df_3['Employee'].str.contains('ski')]
```

## 4. Calculations based on data from the dataframes

Let's assume that we are interested in answering a few questions:
* What is the sum of orders for `Polska`?
* How many orders there are for `Niemcy`?
* What is the average value for an order?

```python
# first we need to select proper data
q_1 = df_3[df_3['Country'] == 'Polska']
q_1
```

```python
# now we can apply a function on a given data or a subset
# this result is not what we were thinking of
q_1.sum()

# now it looks better
q_1['Order_value'].sum()
```

```python
# we could aplly all of the steps in one line - like a chain of functions executed from left to right (more or less)
df_3[df_3['Country'] == 'Polska']['Order_value'].sum()
```

To know the functions, their parameters and values being returned we need to know the API (Application Programming Interface) of pandas DataFrame.
