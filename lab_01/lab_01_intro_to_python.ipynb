Next let's introduce some more commonly used Python data types.

````python
# This is a one line comment - this is not being interpreted, just a plain text.
# integer type (int) - integral numbers (no decimal part), limit? Only computer memory.
# Why the name is num_of_students no num of students?
# Can I use other conventions of naming variables?
# Are there limitations or best practice?
num_of_students = 10

# floating point number
pi = 3.14

# sequence types

# list of values (mutable - values can be changed after declaring a list)
vals = ['John', 'Kate', 10, 0.9999, ['C-3PO', 'Voldemort']]

# tuple of values (immutable - values can not be changed after declaration)
vals_tuple = (1, 2, 3, 'This', 'can', 'be', ['one', 'on', 'one'])

# dict - a dictionary, collection of key - value pairs called also as mapping
dd = {1: 'Monday', 2: 'Tuesday', 'Wednesday': 3}

# a set - like on math lectures, distinct values with set-like operations being already implemented
A = {1, 2, 3}
B = {2, 4, 5}

# display is a built-in function of Jupyter Notebook which acts like print() in Python standard library
# it is used when we want to print out more than just the last refered variable in a single cell (this is a default behaviour)
display(A - B)
display(A & B) # intersection (common part) of sets
display(A.intersection(B))
display(A | B) # sum of sets
````

````python
# let's refer to prevoiusly declared list, tuple and dictionary
display(vals)
display(vals_tuple)
display(dd)
````

````python
# collections (types that can collect/store more than one value inside of it) can be referenced by providing indexes (list, str, tuple) or keys (dict) to refer to
# a specific value being placed at a position (index) or asociated with a name (key in dictionary)
display(vals[0])
display(vals_tuple[1])
display(dd['Wednesday'])
display(name[3])
````

````python
# as mentioned before some of the types are immutable
# trying to replace first letter
name[0] = 'R'
````

````python
# there is also one more built-in sequence type, function range
# documentation for this function looks as follows
print(range.__doc__)
````

````python
# below, there are a few examples of calling range function with different parameters
range(5) # stop = 5, produces a sequence of integers: 0, 1, 2, 3, 4
range(3, 6) #  start = 3, stop = 6, sequence: 3, 4, 5
range(10, 2) # start = 10, stop = 2, but default step is 1, so it produces an empty sequence
range(10, 0, -2) # start = 10, stop = 0, step = -2, sequence: 10, 8, 6, 4, 2
````

````python
# let's try to display those numbers
range(5)
````

````python
# You may ask why I can't see those numbers? Just the function name itself and parameters which were used to call it?
# range is a class (piece of code which is a part of object oriented programming paradigm) which offers a creation
# of immutable sequence of integer numbers. It is not a collection of those numbers, it just has all the necessary
# parameters to create them, but not storing them all in memory. 
# We can use range to ITERATE (go through one by one) all the numbers or to use some other Python types and tricks to
# "materialize" those numbers into a container (aka. collection)

# iterating through all range elements
for item in range(5):
    print(item)
````

````python
# for loop is very useful as it provides possibility to repeat certain operation n times, and that operation could be
# everything that can be programmed
# for loop is also called for each loop as it can "do something" or "get each element" of a collection/sequence
# Internal behavior of for loop is a little bit more complicated, but let's leave it for now

# let's assume, that we need those numbers to be stored in a list
# first, we create an empty list with a list literal ([])
my_list = []

# second, we are using for loop and range function to get those numbers
for i in range(5):
    # anf third, we are using append() method (method is a function created specifically for that object inside it's definition class)
    my_list.append(i)

my_list
````

````python
# as a Python programmer, when we want to get all those numbers without any exceptions from that sequence
# we would probably do it with a little bit less code (= effort)
# We can use feature of a list initialization (process of object creation) where we can provide a sequence as an argument
# which then is internally iterated and all elements are added to a newly created list
list(range(5))
````

````python
# we can of course store it in a variable
my_list_2 = list(range(5))
my_list_2
````

````python
# there is also another loop in Python, a while loop
# while loop is used when the number of iterations is not known at the execution time
# example.
num = 0

while num < 5:
    print(num)
    # increment value of num variable by assigning a new value to it incremented by 1
    # it is also called reassigning variable value
    num = num + 1
````

````python
# while loop takes, a logical (might be only True or False) condition (num < 5) and when if the condition is true
# will perform (execute) code which is inside its block (after : and before next line starting from the beginning of it, no an indentation)
# Above example is somehow similar to the one with for loop, but we can define while loop which is theoreticall infinite (which might be
# tricky to stop when implemented incorrectly).

# we can define a condition which is always true like
# while 0 < 1:
# or even explicit
while True:
    # ask user about a number
    guess_number = 7
    users_number = int(input("Guess which number I'm thinking of!\n"))
    if users_number == guess_number:
        break
````

````python
# we can see a few new things in the code snippet above
# if condition: -> most popular condition usage: if ... then, but without 'then' keyword which might be present in other
# programming languages
# it works in the same logic like condition in while loop, if condition is true, code block executes, if false, this
# block of code is skipped (not executed)

# the break keyword is used to instruct loop that it should immediately stop executing more iterations

# input() function block further execution of code and waits for user input (from keyboard by default)
# it will continue executing when a new line characted is provided (obtained by hitting Enter on your keyboard)
# if the provided number value is equal (==) to the guess_number variable value the loop is stopped
````
